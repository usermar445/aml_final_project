```python
import pandas as pd
import io

heloc_train = pd.read_csv(io.BytesIO(uploaded['heloc_train.csv']))
```

```python
df= pd.DataFrame(heloc_train)
```

```python
df_encoded = pd.get_dummies(df, columns=['RiskPerformance'], drop_first=True)
df_encoded["RiskPerformance"] = df_encoded["RiskPerformance"].astype(int)
```

```python
X = df_encoded.drop('RiskPerformance_Good', axis=1)
y= df_encoded['RiskPerformance_Good']
```

```python
from ensemble_tabpfn import EnsembleTabPFN
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=42)
```

```python
clf = EnsembleTabPFN(max_iters=100)
clf.fit(X_train, y_train)
y_hat = clf.predict(y_test)
acc = accuracy_score(y_test, y_hat)
```
